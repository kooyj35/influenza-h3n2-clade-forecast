```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Elegant ASCII Charts for Clade Analysis - FIXED
==============================================

Creates sophisticated bar charts using only Unicode characters.
Fixed indentation errors.

Usage:
    python elegant_ascii_charts_fixed.py
"""

import csv
import os
import sys
from collections import defaultdict, Counter

# ANSI color codes
COLORS = {
    'blue': '\033[94m',
    'green': '\033[92m',
    'yellow': '\033[93m',
    'red': '\033[91m',
    'purple': '\033[95m',
    'cyan': '\033[96m',
    'white': '\033[97m',
    'bold': '\033[1m',
    'end': '\033[0m'
}

# Unicode characters for sophisticated graphics
UNICODE_BLOCKS = [' ', 'â–', 'â–', 'â–', 'â–Œ', 'â–‹', 'â–Š', 'â–‰', 'â–ˆ']
UNICODE_SHADES = ['â–‘', 'â–’', 'â–“', 'â–ˆ']

class ElegantASCIICharts:
    """Create sophisticated ASCII visualizations."""
    
    def __init__(self):
        pass
    
    def colorize(self, text, color):
        """Add color to text."""
        return f"{COLORS.get(color, '')}{text}{COLORS['end']}"
    
    def create_horizontal_bar_chart(self, data, title, subtitle="", max_width=60, 
                                   color_scheme="blue", show_values=True, 
                                   show_percentage=True):
        """Create sophisticated horizontal bar chart."""
        
        if not data:
            return "No data available"
        
        # Sort data by value (descending)
        sorted_data = sorted(data.items(), key=lambda x: x[1], reverse=True)
        
        # Calculate max value for scaling
        max_value = max(data.values())
        
        # Create chart
        result = []
        
        # Header with decorative border
        border = "â•”" + "â•" * 78 + "â•—"
        result.append(border)
        result.append(self.colorize(f"  {title:^76}  ", 'bold'))
        if subtitle:
            result.append(self.colorize(f"  {subtitle:^76}  ", 'white'))
        result.append("â•š" + "â•" * 78 + "â•")
        
        # Column headers
        result.append(f"\n{'Rank':>4} â”‚ {'Clade':<15} â”‚ {'Count':<8} â”‚ {'Bar Chart':<{max_width}} â”‚ {'%':<6}")
        result.append("â”€" * 80)
        
        # Data rows
        for i, (label, value) in enumerate(sorted_data, 1):
            # Calculate bar length and percentage
            bar_length = int((value / max_value) * max_width)
            percentage = (value / sum(data.values())) * 100
            
            # Create gradient bar using Unicode blocks
            gradient_bar = ""
            full_blocks = bar_length // len(UNICODE_BLOCKS)
            remainder = bar_length % len(UNICODE_BLOCKS)
            
            # Color based on scheme
            if color_scheme == "rainbow":
                colors = ['\033[91m', '\033[93m', '\033[92m', '\033[96m', '\033[94m']
            elif color_scheme == "red":
                colors = ['\033[91m', '\033[31m', '\033[35m']
            else:  # blue default
                colors = ['\033[94m', '\033[96m', '\033[36m']
            
            for j in range(full_blocks):
                color_index = min(j // 3, len(colors) - 1)
                gradient_bar += colors[color_index] + UNICODE_BLOCKS[-1] * len(UNICODE_BLOCKS)
            
            if remainder > 0:
                gradient_bar += colors[min(full_blocks // 3, len(colors) - 1)] + UNICODE_BLOCKS[remainder]
            
            gradient_bar += COLORS['end']
            
            # Format row
            rank = f"{i:2d}"
            clade_name = f"{label:<15}"
            count = f"{value:>8,}"
            percent = f"{percentage:>5.1f}%"
            
            row = f"{rank} â”‚ {clade_name} â”‚ {count} â”‚ {gradient_bar:<60} â”‚ {percent}"
            result.append(row)
        
        # Footer
        result.append("â”€" * 80)
        
        # Summary statistics
        total = sum(data.values())
        mean_val = total / len(data)
        
        summary_stats = f"\nğŸ“Š Summary: {len(data)} clades, {total:,} total samples, avg: {mean_val:.1f}"
        result.append(self.colorize(summary_stats, 'cyan'))
        
        return "\n".join(result)
    
    def create_advanced_bar_chart(self, data, title, subtitle=""):
        """Create advanced bar chart with Unicode box drawing."""
        result = []
        
        # Header with decorative border
        border = "â•”" + "â•" * 78 + "â•—"
        result.append(border)
        result.append(self.colorize(f"  {title:^76}  ", 'bold'))
        if subtitle:
            result.append(self.colorize(f"  {subtitle:^76}  ", 'white'))
        result.append("â•š" + "â•" * 78 + "â•")
        
        # Sort data
        sorted_items = sorted(data.items(), key=lambda x: x[1], reverse=True)
        
        # Create chart with Unicode box drawing
        result.append("\n" + "â”Œ" + "â”€" * 15 + "â”¬" + "â”€" * 10 + "â”¬" + "â”€" * 50 + "â”")
        result.append("â”‚ Clade Name    â”‚ Count    â”‚ Bar Chart" + " " * 41 + "â”‚")
        result.append("â”œ" + "â”€" * 15 + "â”¼" + "â”€" * 10 + "â”¼" + "â”€" * 50 + "â”¤")
        
        max_value = max(data.values())
        
        for label, value in sorted_items[:20]:  # Top 20
            # Create gradient bar
            bar_length = int((value / max_value) * 45)
            
            # Multi-color gradient
            gradient_bar = ""
            for i in range(0, bar_length, 3):
                color_intensity = min(i / 45, 1)
                if color_intensity < 0.33:
                    color = '\033[94m'  # Light blue
                elif color_intensity < 0.66:
                    color = '\033[96m'  # Cyan
                else:
                    color = '\033[92m'  # Green
                
                part_length = min(3, bar_length - i)
                gradient_bar += color + "â–ˆ" * part_length
            
            gradient_bar += COLORS['end']
            
            # Format row with Unicode box drawing
            row = f"â”‚ {label:<13} â”‚ {value:>8,} â”‚ {gradient_bar:<45} â”‚"
            result.append(row)
        
        result.append("â””" + "â”€" * 15 + "â”´" + "â”€" * 10 + "â”´" + "â”€" * 50 + "â”˜")
        
        return "\n".join(result)
    
    def create_comparison_chart(self, test_data, val_data, title):
        """Create side-by-side comparison chart."""
        result = []
        
        # Header
        result.append("\n" + "â•" * 100)
        result.append(self.colorize(f"  {title}  ", 'bold'))
        result.append("â•" * 100)
        
        # Get common clades
        common_clades = set(test_data.keys()) & set(val_data.keys())
        
        # Calculate differences
        differences = {}
        for clade in common_clades:
            diff = abs(test_data[clade] - val_data[clade])
            differences[clade] = diff
        
        # Sort by difference
        sorted_diffs = sorted(differences.items(), key=lambda x: x[1], reverse=True)
        
        # Header for comparison
        result.append(f"\n{'Clade':<15} â”‚ {'TEST':<8} â”‚ {'VAL':<8} â”‚ {'Difference':<10} â”‚ {'Comparison':<40}")
        result.append("â”€" * 100)
        
        max_diff = max(differences.values()) if differences else 1
        
        for label, diff in sorted_diffs[:15]:
            test_val = test_data.get(label, 0)
            val_val = val_data.get(label, 0)
            
            # Create comparison bar
            bar_length = int((diff / max_diff) * 35)
            
            if test_val > val_val:
                comparison_bar = self.colorize("â–ˆ" * bar_length, 'red') + " TEST"
            else:
                comparison_bar = self.colorize("â–ˆ" * bar_length, 'blue') + " VAL"
            
            comparison_bar += COLORS['end']
            
            row = f"{label:<15} â”‚ {test_val:>8,} â”‚ {val_val:>8,} â”‚ {diff:>10,} â”‚ {comparison_bar:<40}"
            result.append(row)
        
        result.append("â”€" * 100)
        
        return "\n".join(result)

def read_clade_data():
    """Read clade data from CSV files."""
    clade_data = defaultdict(int)
    test_data = defaultdict(int)
    val_data = defaultdict(int)
    
    try:
        with open('clade_distribution_summary.csv', 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f)
            for row in reader:
                clade = row.get('Clade', '')
                if clade:
                    combined = int(row.get('Combined_Count', 0))
                    test = int(row.get('TEST_Count', 0))
                    val = int(row.get('VAL_Count', 0))
                    
                    clade_data[clade] = combined
                    test_data[clade] = test
                    val_data[clade] = val
        
        return clade_data, test_data, val_data
        
    except FileNotFoundError:
        print("âŒ clade_distribution_summary.csv not found")
        return None, None, None
    except Exception as e:
        print(f"âŒ Error reading data: {str(e)}")
        return None, None, None

def create_text_summary(clade_data):
    """Create detailed text summary."""
    print("\n" + "â•" * 60)
    print("ğŸ“‹ DETAILED ANALYSIS SUMMARY")
    print("â•" * 60)
    
    total_samples = sum(clade_data.values())
    unique_clades = len(clade_data)
    
    # Top clades
    top_clades = sorted(clade_data.items(), key=lambda x: x[1], reverse=True)[:10]
    
    print(f"""
ğŸ“Š OVERALL STATISTICS
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
  Total Samples: {total_samples:,}
  Unique Clades: {unique_clades:,}
  Average per Clade: {total_samples/unique_clades:.1f}
  
ğŸ† TOP 10 CLADES
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
""")
    
    for i, (clade, count) in enumerate(top_clades, 1):
        percentage = (count / total_samples) * 100
        print(f"  {i:2d}. {clade:15s} â”‚ {count:5d} samples â”‚ {percentage:6.2f}%")
    
    # Diversity metrics
    if len(top_clades) > 4:
        top_5_total = sum(count for _, count in top_clades[:5])
        top_5_percentage = (top_5_total / total_samples) * 100
        
        print(f"""
ğŸ“ˆ DIVERSITY ANALYSIS
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
  Top 5 clades represent: {top_5_percentage:5.1f}% of all samples
  Clade evenness index: {1 - sum((count/total_samples)**2 for count in clade_data.values()):6.3f}
""")

def main():
    """Main function."""
    print("ğŸš€ Starting Elegant ASCII Visualization")
    print("="*80)
    print("Creating sophisticated charts without any external libraries!")
    print("="*80)
    
    # Read data
    clade_data, test_data, val_data = read_clade_data()
    
    if not clade_data:
        print("âŒ No data available")
        return
    
    # Create visualizer
    visualizer = ElegantASCIICharts()
    
    print(f"âœ… Data loaded: {len(clade_data)} clades, {sum(clade_data.values())} total samples")
    
    # 1. Advanced horizontal bar chart
    print(visualizer.create_advanced_bar_chart(
        dict(sorted(clade_data.items(), key=lambda x: x[1], reverse=True)[:15]),
        "ğŸ† TOP 15 MOST COMMON CLADES",
        "Distribution across Korea, China, and Japan"
    ))
    
    # 2. Horizontal bar chart with gradient
    print(visualizer.create_horizontal_bar_chart(
        dict(sorted(clade_data.items(), key=lambda x: x[1], reverse=True)[:12]),
        "ğŸ“Š CLADE ABUNDANCE RANKING",
        "Gradient visualization with color coding",
        max_width=50,
        color_scheme="rainbow"
    ))
    
    # 3. TEST vs VAL comparison
    if test_data and val_data:
        print(visualizer.create_comparison_chart(
            test_data, val_data,
            "ğŸ” TEST vs VAL: DETAILED COMPARISON"
        ))
    
    # 4. Text summary
    create_text_summary(clade_data)
    
    print("\n" + "â•"*80)
    print("âœ¨ All visualizations completed!")
    print("Check the elegant ASCII charts above!")
    print("â•"*80)

if __name__ == "__main__":
    main()
```
